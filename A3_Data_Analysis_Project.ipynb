## The Data

Water quality measurements for 3276 distinct water bodies are contained in the water potability.csv file downloaded from [Kaggle](https://www.kaggle.com/adityakadiwal/water-potability).
Water quality refers to the chemical, physical, and biological properties of water as determined by its intended use. It's most commonly used to refer to a set of criteria against which compliance, which is usually accomplished by water treatment, may be measured. The most popular standards for monitoring and assessing water quality express ecosystem health, human contact safety, and the state of drinking water. Water quality has a big influence on water supply, and it frequently decides which alternatives are available.
The dataset mentioned above contains information for 9 different features whose measures are essential to determine the quality of water. The feature ‘Potability’ is the class label here and tells us whether the corresponding sample of water is safe or consumption or not. Details on all the features are mentioned below. 

**1.	pH value:** The pH of water is an essential metric in determining its acid–base balance. It may also be used to determine if the water is acidic or alkaline. The pH maximum acceptable limit has been set at 6.5 to 8.5 by the WHO. The present investigation ranged from 6.52 to 6.83, which is within WHO guidelines.

**2.	Hardness:** Calcium and magnesium salts are the major causes of hardness. These salts are dissolved in water when it passes through geologic formations. The amount of time water spends in contact with hardness-producing material influences the amount of hardness in raw water. The capacity of water to precipitate soap induced by Calcium and Magnesium was previously characterised as hardness.

**3.	Solids (Total dissolved solids - TDS):** Potassium, calcium, sodium, bicarbonates, chlorides, magnesium, sulphates, and other inorganic and organic minerals or salts may all be dissolved in water. These minerals gave the water an unpleasant flavour and a diluted coloring. This is an essential criterion for water use. The presence of a high TDS value in water implies that it is heavily mineralized. Desirable limit for TDS is 500 mg/l and maximum limit is 1000 mg/l which prescribed for drinking purpose.

**4.	Chloramines:** The main disinfectants used in public water systems are chlorine and chloramine. When ammonia is added to chlorine to purify drinking water, chloramines are produced. Chlorine levels in drinking water up to 4 milligrammes per litre (mg/L or 4 parts per million (ppm)) are deemed safe.

**5.	Sulfate:** Minerals, soil, and rocks all include sulphates, which are naturally occurring compounds. They can be found in the environment, groundwater, plants, and food. Sulfate's main commercial application is in the chemical industry. Seawater contains around 2,700 milligrammes per litre (mg/L) of sulphate. In most freshwater sources, it varies from 3 to 30 mg/L, however considerably greater concentrations (1000 m) are possible.

**6.	Conductivity:** Pure water is an excellent insulator rather than a strong conductor of electric current. The electrical conductivity of water improves as ions concentration rises. The electrical conductivity of water is usually determined by the amount of dissolved particles in it. Electrical conductivity (EC) is a measurement of a solution's ionic process, which allows it to transfer current. As per WHO standards, EC value should not exceeded 400 μS/cm.

**7.	Organic_carbon:** Total Organic Carbon (TOC) in source waters originates from both natural and manmade sources. The total quantity of carbon in organic compounds in pure water is measured by TOC. 2 mg/L TOC in treated / drinking water, and 4 mg/Lit in source water used for treatment, according to the US EPA.

**8.	Trihalomethanes:** THMs are compounds that can be detected in chlorine-treated water. THM levels in drinking water vary depending on the quantity of organic matter in the water, the amount of chlorine needed to treat the water, and the temperature of the water being treated. THM levels in drinking water up to 80 ppm are deemed safe.

**9.	Turbidity:** The amount of solid materials in the suspended state determines the turbidity of water. It is a test that determines the quality of waste discharge in terms of colloidal particles by measuring the light emitting characteristics of water. Wondo Genet Campus's mean turbidity value (0.98 NTU) is lower than the WHO's recommended threshold of 5.00 NTU.

**10.	Potability:** Indicates whether or not water is fit for human consumption, with 1 denoting drinkable and 0 denoting unfit for human consumption.

## Data Pre-processing -
In these sections of codes, we prepare the data for modelling.

The following code cell checks for missing values in the dataset.

```python
water_df.isnull().sum()
```

It can be seen from the output in the previous cell that, the feaures - ph, Sulfate, Trihalomethanes contain missing data in numbers of 491, 781 and 162 respectively. These will be handled in the subsequent code cells.

In the below code cell, we check for the median values of the three feature mentioned above with missing values for non-potable kind of water.

```python
water_df[water_df['Potability']==0][['ph','Sulfate','Trihalomethanes']].median()
```

In the below code cell, we check for the median values of the three feature mentioned above with missing values for potable kind of water.

```python
water_df[water_df['Potability']==1][['ph','Sulfate','Trihalomethanes']].median()
```

It can be observed from the previous two outputs above that the median values for these three features for both the classes are nearly the same. Hence we would impute the missing values for these features with the overall median of these features repectively.

```python
water_df["ph"].fillna(value = water_df["ph"].median(), inplace = True)
water_df["Sulfate"].fillna(value = water_df["Sulfate"].median(), inplace = True)
water_df["Trihalomethanes"].fillna(value = water_df["Trihalomethanes"].median(), inplace = True)
```

The following code cell performs a sanity check to ensure all the missing values have been handled.

```python
water_df.isnull().sum()
```

Now that the missing values have been taken care of, we would prepare the data for modelling. For this purpose we would split our dataset into training and test sets which is done in the following cell.

```python
X = water_df.drop('Potability',axis=1).values
y = water_df['Potability'].values
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

In the following code cell, we standardize the data in our dataset. Since all the features are numerical, we use the StandardScalar() function to perform the same.

```python
scaler = StandardScaler()
scaler.fit(X_train)
X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)
```

## Classification -
Now that our data set is ready for modelling, in these sections of codes, we build four different classifiers and run those classifiers on our dataset and note the accuracy results.

### Decision Tree-
The following code cell creates a decision tree classifier, fits the model into the training data, makes prediction in test data and predicts the accuracy.

```python
DT_model = DecisionTreeClassifier(max_depth=4, random_state=42)# Creating the model
DT_model.fit(X_train, y_train) # Training the model
pred_dt = DT_model.predict(X_test) # making predictions
DT_acc= accuracy_score(y_test, pred_dt) # predicting the accuracy on test data
print(DT_acc)
```

The following code cell prints the classification report from the above execution of the decision tree classfier.

```python
print(classification_report(y_test, pred_dt))
```

### Support Vector Machine (SVM) -
The following code cell creates a SVM classifier, fits the model into the training data, makes prediction in test data and predicts the accuracy.

```python
SVM_model = SVC(kernel='rbf', random_state= 42)
SVM_model.fit(X_train, y_train)
SVM_pred = SVM_model.predict(X_test)
SVM_acc = accuracy_score(y_test, SVM_pred)
print(SVM_acc)
```

The following code cell prints the classification report from the above execution of the SVM classfier.

```python
print(classification_report(y_test, SVM_pred))
```

### Random Forest -
The following code cell creates a Random Forest classifier, fits the model into the training data, makes prediction in test data and predicts the accuracy.

```python
RF_model = RandomForestClassifier(n_estimators=300,min_samples_leaf=0.16, random_state=42)
RF_model.fit(X_train, y_train)
RF_pred = RF_model.predict(X_test)
RF_acc =  accuracy_score(y_test, RF_pred)
print(RF_acc)
```

The following code cell prints the classification report from the above execution of the Random forest classfier.

```python
print(classification_report(y_test, RF_pred))
```

### K Nearest Neighbors -
The following code cell creates a K nearest neighbours classifier, fits the model into the training data, makes prediction in test data and predicts the accuracy.

```python
KN_model = KNeighborsClassifier(n_neighbors=9, leaf_size=20)
KN_model.fit(X_train, y_train)
KN_pred = KN_model.predict(X_test)
KN_acc = accuracy_score(y_test, KN_pred)
print(KN_acc)
```

The following code cell prints the classification report from the above execution of the K-neighbor classfier.

```python
print(classification_report(y_test, KN_pred))
```

SVM maps training examples to points in space in order to widen the distance between the two categories as much as possible. New instances are then mapped into the same space and classified according to which side of the gap they land on.
